#Imbalanced Dataset:
A dataset with an unequal class distribution.

```python
import numpy as np
import pandas as pd
```

```python
credit_card_data = pd.read_csv('/content/credit_data.csv')
```

```python
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
credit_card_data.shape
```

```python
#determine the distribution of two classes
credit_card_data['Class'].value_counts()
```

This is Highly imbalanced Dataset

0 ---> Legit transaction

1 ---> Fraudulent Transactions

```python
#separating
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

#Under-sampling

Building a sample dataset containing similar distribution of Legit & Fraudulent Transactions

```python
legit_sample = legit.sample(n=88)
```

```python
print(legit_sample.shape)
```

#Concatenate two datadrames

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```
